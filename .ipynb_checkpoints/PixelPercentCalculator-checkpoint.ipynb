```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def show (img):
    plt.imshow(img)

filepath=r'C:\Users\kyohe\Solid_red.svg.png'

# 画像を読み込む
im2=Image.open(filepath)
show(im2)
image=np.asarray(im2)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 特定の色の範囲を定義 (例: 赤色)
lower_red = np.array([0, 0, 100])
upper_red = np.array([0, 0, 200])

# マスクを作成
mask = cv2.inRange(image, lower_red, upper_red)
show(mask)

# 特定の色のピクセル数をカウント
red_pixels = cv2.countNonZero(mask)

# 全体のピクセル数を取得
total_pixels = image.shape[0] * image.shape[1]

# 割合を計算
percentage = (red_pixels / total_pixels) * 100

print(f'赤色のピクセルが占める割合: {percentage:.2f}%')
```
